Las diferentes representaciones de Fourier se utilizan para analizar señales en el dominio de la frecuencia. Aquí te explico las semejanzas y diferencias entre ellas:

**Series de Fourier (Exponencial, Trigonométrica y Compacta):**

***Semejanzas:*** Todas representan señales periódicas en el dominio del tiempo como una suma de componentes sinusoidales (o exponenciales complejas) en el dominio de la frecuencia. El espectro es discreto, es decir, solo existen frecuencias específicas en el espectro.

***Diferencias:***

**Trigonométrica:** Utiliza senos y cosenos.

**Exponencial:** Utiliza exponenciales complejas (relacionadas con senos y cosenos a través de la fórmula de Euler). Es más conveniente para análisis matemático.

**Compacta:** Es una forma simplificada de la trigonométrica o exponencial que expresa la amplitud y fase de cada componente.

**Transformada de Fourier (FT):**

***Semejanzas:*** Analiza señales en el dominio de la frecuencia.
Diferencias: Se aplica a señales no periódicas en el dominio del tiempo. El espectro de la FT es continuo, lo que significa que existen componentes en todas las frecuencias posibles. La señal en el tiempo es continua.

**Transformada de Fourier en Tiempo Discreto (DTFT):**

***Semejanzas:*** Analiza señales en el dominio de la frecuencia.

***Diferencias:*** Se aplica a señales discretas en el dominio del tiempo (muestreadas). El espectro de la DTFT es continuo y periódico.

**Transformada Discreta de Fourier (DFT):**

***Semejanza:*** Analiza señales en el dominio de la frecuencia.

***Diferencias:*** Se aplica a señales discretas y finitas en el dominio del tiempo. El espectro de la DFT es discreto y finito. Es la versión computacional de la DTFT.

¿Qué es? La FFT es un algoritmo eficiente para calcular la Transformada Discreta de Fourier (DFT) y su inversa. No es una transformada diferente, sino una forma rápida de calcular la DFT.
Utilidad: La DFT tiene un costo computacional elevado, especialmente para señales largas. La FFT reduce significativamente este costo, haciendo posible el análisis de frecuencia de señales en tiempo real o en sistemas con recursos limitados.
Detalle del Algoritmo: La FFT se basa en el principio de "divide y vencerás". Descompone el cálculo de una DFT de tamaño N en el cálculo de DFTs más pequeñas. El algoritmo más común es el de Cooley-Tukey, que funciona de manera más eficiente cuando el tamaño N es una potencia de 2. Divide la secuencia de entrada en secuencias pares e impares y calcula sus DFTs por separado, combinando luego los resultados para obtener la DFT completa.
Costo Computacional:
DFT (directa): El costo computacional de calcular una DFT directamente es proporcional a $N^2$ (donde N es el número de puntos en la señal). Esto significa que si duplicas el tamaño de la señal, el tiempo de cálculo se cuadruplica.
FFT: El costo computacional de la FFT es proporcional a $\text{N log}_2(\text{N})$. Esto es significativamente menor que $N^2$, especialmente para valores grandes de N. Si duplicas el tamaño de la señal, el tiempo de cálculo aumenta mucho menos que al usar la DFT directa.

En resumen, las representaciones de Fourier nos permiten entender el contenido de frecuencia de una señal. La elección de la representación adecuada depende de si la señal es continua o discreta en el tiempo, y si es periódica o no. La FFT es una herramienta crucial para la implementación práctica de la DFT debido a su eficiencia computacional.

### Transformada de Fourier:

$$
X(\omega) = \int_{-\infty}^{\infty} x(t) e^{-j\omega t} dt
$$



###a) $e^{-a|t|}, a∈\mathbb{R}^+$;

La definición de \( |t| \) es:

$$
|t| =
\begin{cases}
t & \text{si } t \geq 0 \\
-t & \text{si } t < 0
\end{cases}
$$

Entonces:

$$
x(t) =
\begin{cases}
e^{-a t} & t \geq 0 \\
e^{a t} & t < 0
\end{cases}
$$

Transformada:

$$
X(\omega) = \int_{-\infty}^{0} e^{a t} e^{-j\omega t} dt + \int_{0}^{\infty} e^{-a t} e^{-j\omega t} dt
$$

$$
= \int_{-\infty}^{0} e^{(a - j\omega)t} dt + \int_{0}^{\infty} e^{-(a + j\omega)t} dt
$$

Resolviendo:

$$
= \frac{1}{a - j\omega} + \frac{1}{a + j\omega} = \frac{2a}{a^2 + \omega^2}
$$

---

###b) $cos(w_ct), w_c∈\mathbb{R}$;

Usando identidades:

$$
\cos(\omega_c t) = \frac{1}{2}(e^{j\omega_c t} + e^{-j\omega_c t})
$$

Transformada:

$$
X(\omega) = \int_{-\infty}^{\infty} \cos(\omega_c t) e^{-j\omega t} dt
= \pi \left[ \delta(\omega - \omega_c) + \delta(\omega + \omega_c) \right]
$$

---

###c) $sin(w_st), w_s∈\mathbb{R}$;

Usamos:

$$
\sin(\omega_s t) = \frac{1}{2j}(e^{j\omega_s t} - e^{-j\omega_s t})
$$

Transformada:

$$
X(\omega) = \int_{-\infty}^{\infty} \sin(\omega_s t) e^{-j\omega t} dt
= j\pi \left[ \delta(\omega - \omega_s) - \delta(\omega + \omega_s) \right]
$$

---

###d) $f(t)cos(w_ct), w_c∈R,f(t)∈\mathbb{R},\mathbb{C}$;

$$
x(t) = \frac{1}{2}f(t)\left(e^{j\omega_c t} + e^{-j\omega_c t}\right)
$$

Transformada:

$$
X(\omega) = \frac{1}{2}F(\omega - \omega_c) + \frac{1}{2}F(\omega + \omega_c)
$$

###e) $e^{-a|t|^2}, a∈\mathbb{R}^+$.

$$
X(\omega) = \int_{-\infty}^{\infty} e^{-a t^2} e^{-j\omega t} dt = \sqrt{\frac{\pi}{a}} e^{-\frac{\omega^2}{4a}}

###f) $Arect_d(t), A, d ∈ \mathbb{R}$.

\text{rect}_d(t) =

0 & \text{en otro caso}
\end{cases}

$$
X(\omega) = A \int_{-\frac{d}{2}}^{\frac{d}{2}} e^{-j\omega t} dt = A d \cdot \text{sinc}\left( \frac{\omega d}{2\pi} \right)
$$

Donde:

$$
\text{sinc}(x) = \frac{\sin(\pi x)}{\pi x}

# Modulación por Amplitud (AM)

La modulación por amplitud (AM) consiste en variar la amplitud de una portadora sinusoidal según una señal mensaje. En detección coherente, se dispone de una portadora local perfectamente sincronizada en amplitud, fase y frecuencia con la portadora original, lo que permite recuperar la señal mensaje con alta fidelidad.

## Ecuación de la señal AM

La señal modulada se expresa como:
$$
s(t) = [1 + m \cdot m(t)] \cdot \cos(2\pi f_c t)
$$
donde:
- $s(t)$: señal modulada en AM.
- $m$: índice de modulación, típicamente $0 \le m \le 1$ para evitar sobremodulación.
- $m(t)$: señal mensaje (normalizada o centrada según convenga).
- $f_c$: frecuencia de la portadora en Hz.

Para que la envolvente $1 + m \, m(t)$ no se haga negativa, se suele elegir $|m(t)| \le 1$ y $m \le 1$. Si $m>1$ o si $m(t)$ excede ±1, aparecerá sobremodulación y distorsión de la envolvente.

## Cálculo correcto del espectro (FFT) unilaterial

Para graficar el espectro de frecuencia en una FFT de una señal real, usamos `rfft` y normalizamos adecuadamente la magnitud. El procedimiento genérico es:

1. Sea $x[n]$ la señal muestreada en tiempo discreto, de longitud $N$.
2. Calculamos $X[k] = \mathrm{rfft}(x)$, que devuelve $N/2+1$ muestras (componente DC, positivas hasta Nyquist).
3. Construimos la magnitud de un solo lado (single-sided spectrum) como:
   - $P2[k] = \frac{|X[k]|}{N}, \quad k=0,\dots,N/2$
   - Luego, para conservar energía en la representación de amplitud:
     - $P1[0] = P2[0] \quad(\text{componente DC}),$
     - $P1[k] = 2\cdot P2[k],\quad k=1,\dots, N/2 - 1,$
     - Si existe componente Nyquist ($k = N/2$ cuando N es par), normalmente no se duplica:
       $$P1[N/2] = P2[N/2].$$
4. El vector de frecuencias es `freqs = rfftfreq(N, 1/fs)`, va de 0 a fs/2.

De esta forma, al graficar `P1` vs `freqs`, obtenemos la magnitud adecuada de cada componente de frecuencia.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq


# Parámetros ajustables
fc = 1000       # Frecuencia de la portadora en Hz
fs = 16800      # Frecuencia de muestreo en Hz. Se usa una frecuencia 16 veces mayor a la frecuencia máxima (1050)
duration = 0.1    # Duración total de la señal en segundos (p.ej. 0.1 s para buena resolución)
modulation_index = 0.8  # Índice de modulación m (0 <= m <= 1 para evitar sobremodulación)
# Parámetros para el pulso rectangular:
pulse_start = 0.02   # tiempo de inicio del pulso en s (dentro de [0, duration])
pulse_end   = 0.08   # tiempo de fin del pulso en s (<= duration)
# Parámetro para el coseno mensaje:
fm = 50.0  # Frecuencia del coseno mensaje en Hz


# Vector de tiempo
N = int(fs * duration)  # número de muestras
t = np.linspace(0, duration, N, endpoint=False)  # vector de tiempo

# Señal mensaje: pulso rectangular único
# Definimos m_pulse(t) = 1 en [pulse_start, pulse_end), 0 en otro caso.
pulse_message = np.zeros_like(t)
mask = (t >= pulse_start) & (t < pulse_end)
pulse_message[mask] = 1.0
# Para evitar que m(t) exceda ±1, aquí está entre 0 y 1, y con modulation_index <= 1,
# la envolvente 1 + m*m(t) varía entre 1 y 1+m sin volverse negativa.


# Señal mensaje: coseno
cosine_message = np.cos(2 * np.pi * fm * t)
# Aquí m(t) está entre -1 y +1, así que con modulation_index <= 1 la envolvente 1 + m*m(t) ≥ 0.


# Señales moduladas AM
# s(t) = [1 + m * m(t)] * cos(2π f_c t)
carrier = np.cos(2 * np.pi * fc * t)
s_pulse = (1.0 + modulation_index * pulse_message) * carrier
s_cos   = (1.0 + modulation_index * cosine_message) * carrier


# Cálculo de FFT normalizada (single-sided)
def single_sided_spectrum(x, fs):
    """
    Calcula la FFT uniliteral normalizada para una señal real x:
    Retorna (freqs, P1) donde:
    - freqs: vector de frecuencias de 0 a fs/2 de longitud N/2+1
    - P1: magnitud normalizada single-sided, considerando duplicar componentes (excepto DC y Nyquist).
    """
    N = len(x)
    # FFT de una señal real:
    X = rfft(x)
    P2 = np.abs(X) / N  # magnitud de dos lados dividida por N
    P1 = P2.copy()
    # duplicar componentes excepto DC (índice 0) y quizá Nyquist (último índice cuando N es par)
    if N % 2 == 0:
        # hay componente Nyquist en P2[-1], no duplicar:
        P1[1:-1] = 2 * P2[1:-1]
        # P1[0] y P1[-1] se mantienen sin duplicar
    else:
        # si no hay componente Nyquist exacta, duplicar todos excepto DC:
        P1[1:] = 2 * P2[1:]
    freqs = rfftfreq(N, d=1.0/fs)
    return freqs, P1

# Obtener espectros
freqs_pulse, P1_pulse = single_sided_spectrum(s_pulse, fs)
freqs_cos,   P1_cos   = single_sided_spectrum(s_cos, fs)
# (Ambos freqs_pulse y freqs_cos son iguales, pues usan misma fs y N.)


# Graficar
plt.figure(figsize=(12, 8))

# 1) Dominio del tiempo: mostrar las formas de onda moduladas superpuestas,
#    pero acotar el eje x a una ventana corta para ver detalle de la envolvente y la portadora.
t_window = 0.005  # ventana de 5 ms para visualización
plt.subplot(2, 1, 1)
plt.plot(t, s_pulse, label='AM con pulso rectangular', alpha=0.7)
plt.plot(t, s_cos,   label='AM con coseno (50 Hz)', alpha=0.7)
plt.xlim(0, t_window)
plt.title("Señales AM Moduladas en el dominio del tiempo (ventana de {:.1f} ms)".format(t_window*1000))
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.legend()
plt.grid()

# 2) Dominio de la frecuencia: espectros normalizados single-sided
plt.subplot(2, 1, 2)
plt.plot(freqs_pulse, P1_pulse, label='Espectro AM con pulso rectangular', alpha=0.7)
plt.plot(freqs_cos,   P1_cos,   label='Espectro AM con coseno (50 Hz)', alpha=0.7)
plt.title("Espectro de Frecuencia (AM) - Single-sided")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Magnitud normalizada")
plt.xlim(0, fs/2)
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
